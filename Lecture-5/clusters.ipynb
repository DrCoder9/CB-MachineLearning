```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean1 = [4,4]
covar1 = [[1,1], 
          [1,5]]

mean2 = [-1,-1]
covar2 = [[1,0], 
          [0,1]]

d1 = np.random.multivariate_normal(mean1, covar1, 1000)
d2 = np.random.multivariate_normal(mean2, covar2, 1000)

d1.shape
d1

plt.scatter(d1[:,0], d1[:,1])

# The variance-covariance matrix has the following structure:
# [ 𝑣𝑎𝑟(𝑥), 𝑐𝑜𝑣(𝑥,𝑦)
#  𝑐𝑜𝑣(𝑥,𝑦), 𝑣𝑎𝑟(𝑦) ]

# “Covariance” indicates the direction of the linear relationship between variables. 
# “Correlation” on the other hand measures both the strength and direction of the linear relationship between two variables

# https://towardsdatascience.com/let-us-understand-the-correlation-matrix-and-covariance-matrix-d42e6b643c22
# https://datascienceplus.com/understanding-the-covariance-matrix/
```

```python
d1[:, 0].shape
```

```python
plt.figure()
plt.scatter(d1[:, 0], d1[:, 1], label="cluster-1")
plt.scatter(d2[:, 0], d2[:, 1], label="cluster-2")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel("longitude")
plt.ylabel("latitube")
plt.title("Poore 20 number")
plt.legend()
plt.show()
```

```python
np.var(d1[:, 0])
```

```python
d1.shape
```

```python
X = np.vstack([d1, d2])  # THIS IS NOT X-AXIS, we merged both the points d1 and d2

y = np.ones([len(d1) + len(d2)], dtype=int)
y[len(d2):] = 2
```

```python
# y   # this will be used for coloring, THIS IS NOT Y-AXIS
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y) # c means color is y
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel("longitude")
plt.ylabel("latitube")
plt.title("Poore 20 number")
plt.show()
```

```python
from sklearn.datasets import make_blobs
```
